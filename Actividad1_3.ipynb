```python
#Importar librerías
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('Inmuebles.xlsx',header=1)
df.head()
```

```python
#Aplicar los siguientes filtros
#Filtro A: Eliminar la fila "0" de "Unnamed" y eliminar la columna "0" y la columna "Referencia"
filtro_a=df.drop('Referencia', axis=1)
filtro_a
```

```python
#Los siguientes filtros son sobre filtro_a
#Filtro b: "Fecha_Alta" igual a 2007-04-29 y a 2004-01-02
filtro_b=filtro_a[(filtro_a["Fecha_Alta"]=="2007-04-29") | (df["Fecha_Alta"]=="2004-01-02")]
filtro_b
```

```python
#Filtro c: Todos los registros de "Tipo" parking, local, industrial y piso
filtro_c=filtro_a[(filtro_a["Tipo"]=="Parking") | (df["Tipo"]=="Local") | (df["Tipo"]=="Industrial") | (df["Tipo"]=="Piso")]
filtro_c
```

```python
#Filtro d: Todos los registros con "Provincia" igual a "Girona" y con "Superficie" mayor a 100
filtro_d=filtro_a[(filtro_a["Provincia"]=="Girona") & (df["Superficie"]>100)]
filtro_d
```

```python
#Filtro e: Todos los registros con "Precio Venta" mayor a 1,000,000 o "Vendedor" igual a "Carmen" o "Joaquin"
filtro_e=filtro_a[(filtro_a["Precio Venta"]>1000000) | (df["Vendedor"]=="Carmen") | (df["Vendedor"]=="Joaquin")]
filtro_e
```

```python
#Filtro f: Todos los registros que no sean de "Provincia":"Barcelona"
filtro_f=filtro_a[~(filtro_a["Provincia"]=="Barcelona")]
filtro_f
```

```python
#Comprobación filtro_f: no debe salir Barcelona
np.unique(filtro_f["Provincia"])
```

```python
#Filtro g: Todos los registros que no sean de Operación: alquiler ni Tipo: oficina
filtro_g=filtro_a[~(filtro_a["Operación"]=="Alquiler") & ~(filtro_a["Tipo"]=="Oficina")]
filtro_g
```

```python
#Conversión a archivos csv
filtro_a.to_csv("filtro_a.csv")
filtro_b.to_csv("filtro_b.csv")
filtro_c.to_csv("filtro_c.csv")
filtro_d.to_csv("filtro_d.csv")
filtro_e.to_csv("filtro_e.csv")
filtro_f.to_csv("filtro_f.csv")
filtro_g.to_csv("filtro_g.csv")
```
